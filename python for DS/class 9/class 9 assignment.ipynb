* Create student class where we can create instance if student education must be intermediate or above
               

```python
class Student():
    def __init__(self,sid,name,education):
        if education < 'intermediate' or education >= 'intermediate':
            raise StudenteducationError("Student education must be intermediate or above")
        self.studentid = sid # attribute
        self.name = name
        self.education = education
        
class StudenteducationError(Exception):
    pass
        
s1 = Student(1,"Adil", "intermediate")
print(s1.studentid, s1.name, s1.age)
```

```python
try:
    s1 = Student(1,"adil", "intermediate")
    
except Exception as e:
    print("Error",e)
```
