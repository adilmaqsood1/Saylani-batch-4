* Create Application where we will pass 9 columns values as input then that application will predict and show that result on your GUI 

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import tkinter as tk
from tkinter import ttk

# Generate random data for demonstration
np.random.seed(42)
num_samples = 100
data = {
    f"Feature_{i+1}": np.random.rand(num_samples) for i in range(9)
}
data["Target_Variable"] = 2 * data["Feature_1"] + 3 * data["Feature_2"] + np.random.normal(0, 0.1, num_samples)

# Convert the data dictionary to a DataFrame
df = pd.DataFrame(data)

# Split data into features (X) and target variable (y)
X = df.drop("Target_Variable", axis=1)
y = df["Target_Variable"]

# Build the model
model = LinearRegression()
model.fit(X, y)

def predict():
    # Get the input values from the Entry widgets
    input_values = [float(entry.get()) for entry in entry_widgets]
    
    # Make the prediction using the model
    prediction = model.predict([input_values])
    
    # Display the result in the label widget
    result_label.config(text=f"Prediction: {prediction[0]:.2f}")

# Create the main application window
app = tk.Tk()
app.title("Prediction Application")
app.geometry("400x300")

# Set a custom font
app.option_add("*Font", "Helvetica 12")

# Create Entry widgets for input values
entry_widgets = []
for i in range(9):
    label = ttk.Label(app, text=f"Feature {i+1}:")
    label.grid(row=i, column=0, padx=5, pady=5, sticky="e")
    
    entry = ttk.Entry(app, font=("Helvetica", 12))
    entry.grid(row=i, column=1, padx=5, pady=5)
    entry_widgets.append(entry)

# Create a separator
separator = ttk.Separator(app, orient="horizontal")
separator.grid(row=9, column=0, columnspan=2, pady=10, sticky="ew")

# Create a button to trigger the prediction
predict_button = ttk.Button(app, text="Predict", command=predict)
predict_button.grid(row=10, column=0, columnspan=2, padx=5, pady=10)

# Create a label to display the result
result_label = ttk.Label(app, text="Prediction: N/A", font=("Helvetica", 14, "bold"))
result_label.grid(row=11, column=0, columnspan=2)

# Start the Tkinter event loop
app.mainloop()
```
