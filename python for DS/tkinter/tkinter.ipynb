```python
import tkinter as tk
from tkinter import filedialog
from PIL import Image, ImageEnhance, ImageFilter, ImageTk


def open_image():
    filepath = filedialog.askopenfilename(filetypes=[("Image Files", "*.jpg;*.jpeg;*.png;*.gif;*.bmp")])
    if filepath:
        image = Image.open(filepath)
        image.thumbnail((400, 400))  # Resize the image to fit the canvas
        photo = ImageTk.PhotoImage(image)  # Create a Tkinter-compatible photo image
        canvas.image = photo  # Store the photo image as an attribute of the canvas widget
        canvas.create_image(0, 0, anchor="nw", image=photo)


def save_image():
    if canvas.image:
        filepath = filedialog.asksaveasfilename(defaultextension=".jpg",
                                                filetypes=[("JPEG", "*.jpg"), ("PNG", "*.png"), ("GIF", "*.gif")])
        if filepath:
            canvas.image.save(filepath)


def apply_grayscale():
    if canvas.image:
        image = canvas.image_obj.convert("L")
        photo = ImageTk.PhotoImage(image)
        canvas.image = photo
        canvas.create_image(0, 0, anchor="nw", image=photo)


def apply_sharpen():
    if canvas.image:
        image = canvas.image_obj.filter(ImageFilter.SHARPEN)
        photo = ImageTk.PhotoImage(image)
        canvas.image = photo
        canvas.create_image(0, 0, anchor="nw", image=photo)


def adjust_brightness(value):
    if canvas.image:
        brightness = float(value) / 100.0
        enhancer = ImageEnhance.Brightness(canvas.image_obj)
        image = enhancer.enhance(brightness)
        photo = ImageTk.PhotoImage(image)
        canvas.image = photo
        canvas.create_image(0, 0, anchor="nw", image=photo)


def remove_background():
    # Placeholder for background removal functionality
    pass


# Create the main window
window = tk.Tk()

# Set the window title
window.title("Simple Photoshop")

# Create a canvas to display the image
canvas = tk.Canvas(window, width=400, height=400)
canvas.pack()

# Create buttons for image operations
open_button = tk.Button(window, text="Open Image", command=open_image)
open_button.pack()

save_button = tk.Button(window, text="Save Image", command=save_image)
save_button.pack()

grayscale_button = tk.Button(window, text="Grayscale", command=apply_grayscale)
grayscale_button.pack()

sharpen_button = tk.Button(window, text="Sharpen", command=apply_sharpen)
sharpen_button.pack()

brightness_label = tk.Label(window, text="Brightness:")
brightness_label.pack()

brightness_slider = tk.Scale(window, from_=0, to=200, orient="horizontal", command=adjust_brightness)
brightness_slider.set(100)
brightness_slider.pack()

background_button = tk.Button(window, text="Remove Background", command=remove_background)
background_button.pack()

# Run the main event loop
window.mainloop()
```
