# data types

```python
import numpy as np

arr = np.array([1, 2, 3], dtype=np.float64)                  # Create an array with a specific data type
print(arr.dtype)
```

```python
import numpy as np

# Create an array with the default data type (int)
arr = np.array([1, 2, 3])
print(arr.dtype)
print(arr)

arr = arr.astype(np.float64)                            # Change the data type to float
print(arr.dtype)
print(arr)
```

```python
import numpy as np

# Create an array with the default data type (int)
arr = np.array([1, 2, 3])
print(arr.dtype)                                                       

arr = arr.astype(float)                                   # Convert the data type to float             
print(arr.dtype)
```

```python
dtype = np.int32
size = np.dtype(dtype).itemsize
print(size)
```

```python
# Create an array with the smallest data type that can hold the values
arr = np.array([1, 2, 3], dtype=np.int8)
print(arr.dtype)
```

```python
# Check if two data types are equivalent
dtype1 = np.float32
dtype2 = np.float

equivalent = np.can_cast(dtype1, dtype2)
print(equivalent)
```

```python
# Check if a data type is signed or unsigned
dtype = np.int32

signed = np.issubdtype(dtype, np.signedinteger)
unsigned = np.issubdtype(dtype, np.unsignedinteger)

print("Signed:", signed)
print("Unsigned:", unsigned)
```

```python
# Check if a data type is floating-point
dtype = np.float64

is_float = np.issubdtype(dtype, np.floating)
print(is_float)
```

```python
# Check if a data type is an integer
dtype = np.int32

is_int = np.issubdtype(dtype, np.integer)
print(is_int)
```

```python
# Check if a data type is complex
dtype = np.complex128

is_complex = np.issubdtype(dtype, np.complexfloating)
print(is_complex)
```
