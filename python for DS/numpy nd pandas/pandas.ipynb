```python
import pandas as pd

# Read data from the CSV file
data = pd.read_csv("tested.csv")

# Display the first few rows of the DataFrame
print(data.head())

# Get basic statistics about the DataFrame
print(data.describe())

# Check for missing values
missing_values = data.isnull().sum()
print(missing_values)

# Filter passengers who survived
survived_passengers = data[data['Survived'] == 1]

# Group data by 'Sex' and calculate the number of survivors in each group
survivors_by_sex = survived_passengers.groupby('Sex')['Survived'].sum()
print(survivors_by_sex)

# Group data by 'Pclass' and calculate the average age of passengers in each class
average_age_by_class = data.groupby('Pclass')['Age'].mean()
print(average_age_by_class)

# Create a new column 'FamilySize' by combining 'SibSp' (siblings/spouses) and 'Parch' (parents/children)
data['FamilySize'] = data['SibSp'] + data['Parch']

# Filter passengers who traveled alone (FamilySize = 0)
alone_passengers = data[data['FamilySize'] == 0]

# Calculate the percentage of passengers who survived and traveled alone
percentage_survived_alone = (alone_passengers['Survived'].sum() / len(alone_passengers)) * 100
print(f"Percentage of passengers who survived and traveled alone: {percentage_survived_alone:.2f}%")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Task 1: Data loading and exploration
data = pd.read_csv("tested.csv")

# Display the first few rows of the DataFrame
print(data.head())

# Get basic statistics about the DataFrame
print(data.describe())

# Check for missing values in each column
print(data.isnull().sum())

# Task 2: Handling missing values
# Fill missing values in the 'Age' column with the median age
data['Age'].fillna(data['Age'].median(), inplace=True)

# Fill missing values in the 'Embarked' column with the most common value
data['Embarked'].fillna(data['Embarked'].mode()[0], inplace=True)

# Task 3: Analyzing passenger demographics
# Count the number of male and female passengers
gender_counts = data['Sex'].value_counts()
print(gender_counts)

# Calculate the average age of passengers
average_age = data['Age'].mean()
print(f"Average age of passengers: {average_age:.2f} years")

# Task 4: Analyzing survival rates based on various factors
# Calculate the overall survival rate
overall_survival_rate = data['Survived'].mean() * 100
print(f"Overall survival rate: {overall_survival_rate:.2f}%")

# Calculate survival rates based on passenger class (Pclass)
class_survival_rate = data.groupby('Pclass')['Survived'].mean() * 100
print("Survival rate by passenger class:")
print(class_survival_rate)

# Calculate survival rates based on gender
gender_survival_rate = data.groupby('Sex')['Survived'].mean() * 100
print("Survival rate by gender:")
print(gender_survival_rate)

# Task 5: Visualizing the data
# Plot the survival rate by passenger class
plt.bar(class_survival_rate.index, class_survival_rate.values)
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate (%)")
plt.title("Survival Rate by Passenger Class")
plt.show()

# Plot the survival rate by gender
plt.bar(gender_survival_rate.index, gender_survival_rate.values)
plt.xlabel("Gender")
plt.ylabel("Survival Rate (%)")
plt.title("Survival Rate by Gender")
plt.show()

# Plot a histogram of passenger ages
plt.hist(data['Age'], bins=20)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Passenger Age Distribution")
plt.show()
```
