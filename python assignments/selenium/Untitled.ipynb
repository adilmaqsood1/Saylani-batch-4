```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
import time

# Specify the path to the Chrome driver executable
driver_path = 'C:\\Users\\Lenovo\\Desktop\\c driver\\ to \\Chromedriver.exe'

# Create a new instance of the Chrome driver
driver = webdriver.Chrome(driver_path)

# Navigate to LinkedIn
driver.get('https://www.linkedin.com')

# Find the username and password input fields, and enter your login credentials
try:
    username_input = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CSS_SELECTOR, 'input[name="session_key"]'))
    )
    password_input = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CSS_SELECTOR, 'input[name="session_password"]'))
    )

    username_input.send_keys('adilmaqsood501@gmail.com')  # Replace with your LinkedIn email address
    password_input.send_keys('maxbryant50')  # Replace with your LinkedIn password

    # Submit the login form
    password_input.send_keys(Keys.RETURN)

    # Wait until the search input field is visible
    search_input = WebDriverWait(driver, 10).until(
        EC.visibility_of_element_located((By.CSS_SELECTOR, "input[aria-label='Search']"))
    )

    # Enter the search query for "developers 5 years experience"
    search_input.send_keys('developers 5 years experience')
    search_input.send_keys(Keys.RETURN)

    # Wait until the "People" tab is visible and click on it
    people_tab = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.XPATH, "//button[text()='People']"))
    )
    people_tab.click()

    # Scroll to the bottom of the page to load all search results
    last_height = driver.execute_script("return document.body.scrollHeight")
    while True:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(1)
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height

    # Find the search results again after scrolling
    search_results = WebDriverWait(driver, 10).until(
        EC.presence_of_all_elements_located((By.XPATH, "//li[contains(@class, 'search-result__occluded-item')]"))
    )

    # Create a text file to store the candidate details
    file_path = 'C:\\Users\\Lenovo\\Desktop\\candidates_data.txt'
    with open(file_path, 'w') as file:
        # Extract information for the top 5 search results
        for result in search_results[:5]:
            try:
                # Get the name and headline of the candidate
                name = result.find_element(By.XPATH, ".//span[contains(@class, 'name')]").text
                headline = result.find_element(By.XPATH, ".//p[contains(@class, 'subline-level-1')]").text

                # Write the details to the text file
                file.write(f"Name: {name}\n")
                file.write(f"Headline: {headline}\n")
                file.write('\n')
            except NoSuchElementException:
                continue

    # Close the browser
    driver.quit()

    print(f"Candidate details saved in {C:\\Users\\Lenovo\\Desktop\\candidates_data.txt}.")

except Exception as e:
    print(f"An error occurred: {str(e)}")
    driver.quit()
```
